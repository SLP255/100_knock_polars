```python
import polars as pl
```

```python
# use fully print tables 
pl.Config.set_tbl_cols(-1)
pl.Config.set_tbl_rows(-1)
```

```python
#011
df_customer = pl.read_csv("ref_files/data/customer.csv")
print(
    df_customer.filter(pl.col("customer_id").str.ends_with("1")).head(10)
)
```

```python
# 012
df_store = pl.read_csv("ref_files/data/store.csv")
print(
    df_store.filter(pl.col("address").str.contains("横浜市")).head(10)
)
```

```python
# 013
print(
    df_customer.filter(pl.col("status_cd").str.contains(r"^[A-F]+")).head(10)
)
```

```python
# 014
print(
    df_customer.filter(pl.col("status_cd").str.contains(r"[0-9]$")).head(10)
)
```

```python
# 015
print(
    df_customer.filter(pl.col("status_cd").str.contains(r"^[A-F].+[0-9]$")).head(10)
)
```

```python
# 016
print(
    df_store.filter(pl.col("tel_no").str.contains("^[0-9]{3}-[0-9]{3}-[0-9]{4}$")).head(10)
)
```

```python
# 017
print(
    df_customer.sort(pl.col("birth_day")).head(10)
)
```

```python
# 018
print(
    df_customer.sort(pl.col("birth_day"), reverse=True).head(10)
)
```

```python
# 019
df_receipt = pl.read_csv("ref_files/data/receipt.csv")
print(
    df_receipt.with_columns(
        pl.col("amount").rank("min", reverse=True).alias("rank")
    ).sort("rank").select([
        pl.col("customer_id"),
        pl.col("amount"),
        pl.col("rank"),
    ]).head(10)
)
```

```python
# 020
print(
    df_receipt.with_columns(
        pl.col("amount").rank("ordinal", reverse=True).alias("rank")
    ).sort("rank").select([
        pl.col("customer_id"),
        pl.col("amount"),
        pl.col("rank"),
    ]).head(10)
)
```
