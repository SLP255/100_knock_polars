```python
import polars as pl
```

```python
# use fully print tables 
pl.Config.set_tbl_cols(-1)
pl.Config.set_tbl_rows(-1)
```

```python
# 001
# local_file(ignored)
df_receipt = pl.read_csv("ref_files/data/receipt.csv")
print(df_receipt.head(10))
```

```python
# 002
print(df_receipt.select(["sales_ymd", "customer_id", "product_cd", "amount"]).head(10))
```

```python
# 003
print(df_receipt.select(
    [pl.col("sales_ymd").alias("sales_date"), "customer_id", "product_cd", "amount"]
).head(10))
```

```python
# 004
print(
    df_receipt.select(
        ["sales_ymd", "customer_id", "product_cd", "amount"]
    ).filter(
        pl.col("customer_id") == "CS018205000001"
    )
)
```

```python
# 005
print(
    df_receipt.select(
        ["sales_ymd", "customer_id", "product_cd", "amount"]
    ).filter(
        pl.col("customer_id") == "CS018205000001"
    ).filter(
        pl.col("amount") >= 1000
    )
)
```

```python
# 006
print(
    df_receipt.select(
        ["sales_ymd", "customer_id", "product_cd", "quantity", "amount"]
    ).filter(
        pl.col("customer_id") == "CS018205000001"
    ).filter(
        (pl.col("amount") >= 1000) | (pl.col("quantity") >= 5)
    )
)
```

```python
# 007
print(
    df_receipt.select(
        ["sales_ymd", "customer_id", "product_cd", "amount"]
    ).filter(
        pl.col("customer_id") == "CS018205000001"
    ).filter(
        pl.col("amount").is_between(1000, 2000)
    )
)
```

```python
# 008
print(
    df_receipt.select(
        ["sales_ymd", "customer_id", "product_cd", "amount"]
    ).filter(
        pl.col("customer_id") == "CS018205000001"
    ).filter(
        pl.col("product_cd") != "P071401019"
    )
)
```

```python
#009
df_store = pl.read_csv("ref_files/data/store.csv")
print(
    df_store
    .filter(
        pl.col("prefecture_cd") != 13
    ).filter(
        pl.col("floor_area") <= 900
    )
)
```
